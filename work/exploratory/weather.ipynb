```python
# iterate over df_weather2023 columns not the date one
df_temperature = pd.DataFrame()
for col in df_weather2023.columns:
    if col == "Date":
        continue
    new_df = df_weather2023[['Date', col]].copy()
    new_df['Hour'] = col.split('_')[1].split('h')[0]
    meassure_value = col.split('_')[0]
    if meassure_value == 'R':
        break
    new_df['Temperature'] = new_df[col]
    new_df.drop(columns=[col], inplace=True)  # Drop the original column
    df_temperature = pd.concat([df_temperature, new_df])

df_temperature
```

```python
# iterate over df_weather2023 columns not the date one
df_rainfall = pd.DataFrame()
for col in df_weather2023.columns:
    if col == "Date":
        continue
    new_df = df_weather2023[['Date', col]].copy()
    new_df['Hour'] = col.split('_')[1].split('h')[0]
    meassure_value = col.split('_')[0]
    if meassure_value == 'T':
        continue
    new_df['Rainfall'] = new_df[col]
    new_df.drop(columns=[col], inplace=True)  # Drop the original column
    df_rainfall = pd.concat([df_rainfall, new_df])

df_rainfall
```

```python
#replace nan with 0
df_rainfall.fillna(0, inplace=True)
df_rainfall 
```

```python
df_temperature.isna().sum()
```

```python
#merge temperature and rainfall
df_weather = pd.merge(df_temperature, df_rainfall, on=['Date', 'Hour'])
df_weather
```

```python
# delete rows with missing meassurements temperature="-" and rainfall="-"
df_weather = df_weather[(df_weather['Temperature'] != "-") & (df_weather['Rainfall'] != "-")]
df_weather
```
